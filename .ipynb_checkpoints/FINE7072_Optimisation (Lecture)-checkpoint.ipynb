```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import minimize
%matplotlib inline
```

#### 1. Import data (Style Returns, Benchmark and Manager Returns) 

```python
man_data=pd.read_excel('Optimisation Data.xlsx',sheet_name='Manager Data',index_col=0)
style_data=pd.read_excel('Optimisation Data.xlsx',sheet_name='Style Data',index_col=0)
```

```python
man_data
```

```python
style_ret=style_data.iloc[:,:-1].reindex(index=man_data.index)
bnch_ret=style_data.iloc[:,-1].reindex(index=man_data.index)
```

```python
bnch_ret
```

We want to define weights or factor loadings to styles that best explain our manager returns
$$
    \Sigma(Y - X\beta)^{2} = SSE
$$

$$
    \frac{\partial SSE}{\partial \beta}
$$
where $\beta$ minimises SSE

#### Optimisation to reduce SSE 

```python
style_ret
```

```python
guess_beta=np.ones(len(style_ret.columns))/len(style_ret.columns)
```

```python
guess_beta
```

```python
np.sum((man_data['Sands']-style_ret@guess_beta)**2)
```

```python
def obj_func(Î²,y,X):
    y_hat=X@Î²
    resid=y-y_hat
    sse=np.sum(resid**2)
    return sse 

def rbsa_func(y,X):
    guess_Î²=np.ones(len(X.columns))/len(X.columns)
    cons=[{'type':'eq','fun':lambda x:np.sum(x)-1.0},
            {'type':'ineq','fun':lambda x:x}]
    rbsa_raw=minimize(obj_func,guess_Î²,args=(y,X),method='SLSQP',constraints=cons,tol=0.1e-10).x
    rbsa_ser=pd.Series(np.round(rbsa_raw,6),index=X.columns)
    return rbsa_ser
```

```python
style_ret_adj=style_ret.drop(['Big','Med'],axis=1)
```

```python
man_out=rbsa_func(man_data['Sands'],style_ret_adj)
bnch_out=rbsa_func(bnch_ret,style_ret_adj)
```

```python
fig,ax=plt.subplots(2,1,sharex=True,figsize=(10,5),gridspec_kw={'height_ratios': [2, 1]})
clrs=['red','green','blue','violet','purple','navy','darkorchid','gold','teal']
man_out.plot(kind='bar',color=clrs,edgecolor='k',ax=ax[0])
bnch_out.plot(kind='bar',color=clrs,edgecolor='k',ax=ax[1])
ax[0].set_title("Sands vs MSCI")
```

```python
fig,ax=plt.subplots(1,1,figsize=(12,8))
style_bench=style_ret_adj@man_out
(1+style_bench).cumprod().plot(ax=ax,color='teal',label='Style Benchmark',marker='*')
(1+man_data['Sands']).cumprod().plot(ax=ax,color='seagreen',label='Manager')
(1+bnch_ret).cumprod().plot(ax=ax,color='k',linestyle='--',label='MSCI')
ax.legend()
```
